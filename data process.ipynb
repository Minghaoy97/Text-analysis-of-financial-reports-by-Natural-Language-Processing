```python
import os, sys
## split the pdf file
import PyPDF2
from PIL import Image 
import pytesseract 
import sys 
from pdf2image import convert_from_path 
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe' #target to the tesseract executive file.

path = "D:\Stock 10Ks"
os.chdir(path)
```

```python
def mkdir(path): # make directories
 
    folder = os.path.exists(path)
 
    if not folder:                   
        os.makedirs(path)            
        print (path + " file created")
 
    else:
        print ("file existed")
```

```python
targets = os.listdir(path)
fail = []
for i in range(10, len(targets)):
    os.chdir(path + "\\" + targets[i])
    reports = os.listdir()
    for j in range(len(reports)):
        reports[j] = reports[j].split('.')[0]
    for j in range(len(reports)):
        mkdir(reports[j])  #Firstly, create a directory for single pdf
        pdf_num = 0    #record the number of the separated small pdf, can be seen as their ids
        pdfReader = PyPDF2.PdfFileReader(reports[j] + ".pdf") 
        pdfnums = pdfReader.numPages #record the total number of pages within the pdf
        innumber = 10  #split the big pdf into small pdfs with only 10 pages
        outnums = pdfnums // innumber + 1  #total number of splited pdfs
        
        outfile = reports[j] + ".txt"
        os.chdir(path + "\\" + targets[i] + "\\" + reports[j]) #set the working path to the specific directory of the pdf

        try:   
            for num in range(outnums - 1):  # read each pdf pages
                pdfWriter = PyPDF2.PdfFileWriter()
                for pageNum in range(innumber * num, innumber * (num + 1)):
                    pageObj = pdfReader.getPage(pageNum) # read the 10 pages and write them into another pdf file
                    pdfWriter.addPage(pageObj)
                with open('PDFREADER %s' %num + '.pdf', 'wb') as pdfOutputFile:
                    pdfWriter.write(pdfOutputFile)

            pdfWriter = PyPDF2.PdfFileWriter() # transform each small pdf into jpg images
            for pageNum in range(innumber * (outnums - 1), pdfnums):
                pageObj = pdfReader.getPage(pageNum)
                pdfWriter.addPage(pageObj)
            with open('PDFREADER %s' %(outnums - 1) + '.pdf', 'wb') as pdfOutputFile:
                pdfWriter.write(pdfOutputFile)

            pdfs = os.listdir()  # get the list of splitpdfs
             # output text file

            while pdf_num < len(pdfs): #use pdf_num to point to each pdf file
                PDF_file = 'PDFREADER %s' %pdf_num + '.pdf'
                pages = convert_from_path(PDF_file, 500) 

                image_counter = 1 # Counter to store images of each page of PDF to image 

                # Iterate through all the pages stored above 
                for page in pages: 

                    filename = "page_"+str(image_counter+pdf_num*10)+".jpg"

                    # Save the image of the page in system 
                    page.save(filename, 'JPEG') 
                    print("Page"+ str(image_counter+pdf_num*10)+"transfered")

                    # Increment the counter to update filename 
                    image_counter += 1

                #Recognizing text from the images using OCR 

                # Creating a text file to write the output 
                f = open(outfile, "a", encoding='utf-8') # Note that python need utf-8 to correctly read and write the file
                for j in range(1, image_counter): 

                    filename = "page_"+str(j+pdf_num*10)+".jpg"

                    # Recognize the text as string in image using pytesserct 
                    text = str(((pytesseract.image_to_string(Image.open(filename))))) 

                    text = text.replace('-\n', '')	 

                    # write the processed text to the file. 
                    f.write(text)
                    print("text"+str(j+pdf_num*10)+"transfered")

                # Close the file after writing all the text. 
                f.close()
                pdf_num += 1
        except:
            fail.append(targets[i]) # append failed pdf name to the list
        os.chdir(path + "\\" + targets[i])
```
